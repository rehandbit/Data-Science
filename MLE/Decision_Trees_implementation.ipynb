```python
#one-hot encoding

geography_onehot= pd.get_dummies(df['Geography'],drop_first=True).astype('int')
geography_onehot
```

```python
geography_onehot.head()
```

pd.get_dummies(df.Gender)

```python
gender_onehot=pd.get_dummies(df['Gender'],drop_first=True).astype('int')
```

```python
# df.drop(['Geography','Gender'],axis=1,inplace=True)

df2 = pd.concat([df,geography_onehot,gender_onehot],axis=1)
df2
```

```python
df2.head(50)
```

```python
df2.drop(['Gender','Geography'], axis= 1, inplace = True)
```

```python
df2
```

```python
#x, y split
X=df2.drop('Exited',axis=1)

y=df2.Exited
```

```python
#train test split:

from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
```

#Tuning the hyper parameters

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
parameter_grid = {
    'max_depth': [5,10,15,None],
    'min_samples_split': [2,5,10,15],
    'min_samples_leaf':[10,15,20],
}
```

```python
gridsearch = GridSearchCV(
    estimator=DecisionTreeClassifier(),
    param_grid = parameter_grid,
    scoring='f1',
    cv=5,
    verbose=1)
```

```python
gridsearch.fit(X_train,y_train)
```

```python
gridsearch.best_params_
```

```python
model2=DecisionTreeClassifier(max_depth=5,min_samples_leaf=10,min_samples_split=2)
```

```python
model2.fit(X_train,y_train)
```

```python
y_train_pred=model2.predict(X_train)
y_test_pred=model2.predict(X_test)
```
